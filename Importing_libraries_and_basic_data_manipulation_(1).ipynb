##### Pandas

Pandas is a powerful library for data analysis and manipulation. It offers a high-level data structure called a DataFrame, a two-dimensional labeled data table similar to a spreadsheet.

Key Features:
- DataFrames for organizing and handling labeled data.
- Easy-to-use data loading and cleaning tools.
- Selection, filtering, and grouping operations on DataFrames.
- Merging and joining DataFrames from different sources.

- DataFrames are the heart of data analysis in Pandas. You can create DataFrames from various sources (CSV, Excel, databases) and use them for:
    - Exploring data (descriptive statistics, sorting, filtering).
    - Data cleaning (handling missing values, outliers).
    - Feature engineering (creating new features from existing ones).
    - Data preparation for machine learning models.

#### Loading Data

Data loading is a crucial step in many data analysis and machine learning tasks.

Loading CSV Data with Pandas

```python
import pandas as pd

# CSV data into a DataFrame
df = pd.read_csv('/content/childhealthdata.csv')
df.head(10)
```

```python
df.tail()
```

- In this example, we've loaded a CSV file into a Pandas DataFrame using the read_csv() function.
- The head() method is then used to display the first few rows of the DataFrame, giving us an initial glimpse of the data.

Exploring the Dataset

- It's crucial to understand the structure and characteristics of the dataset you are working with. Let's explore some key attributes and statistics of the loaded DataFrame

Displaying basic information about the DataFrame using info() function

```python
print(df.info())
```

- This line of code provides a quick overview of the DataFrame's structure, helping you understand the data types of each column, the presence of missing values, and the overall memory footprint of the data.

A summary of the numerical columns using desribe()

```python
print(df.describe())
```

- The code describes numerical columns (containing numbers). Categorical columns (containing text or labels) are excluded from the output.
- If a column has a lot of missing values (null values), the count for that column might be significantly lower than the total number of rows in the DataFrame.
- The percentiles (25%, 50%, and 75%) provide insights into the distribution of the data within each column.

Getting Unique values in a specific column using unique()

```python
print(df['education'].unique())
```

- The unique() method is applied to the selected column (df['education']) which returns the unique values within the 'education' column.

Frequency distribution of values in a specific column

```python
print(df['education'].value_counts())
```

- value_counts() is a method applied to the selected column (df['education']). It calculates the frequency or count of each unique value within the 'education' column.

Column operations

We can do mathematical operations between dataframe columns and they'll be applied elementwise across each row.

We can also define new columns as a result of these operations. For example:

```python
df['age_in_days'] = df['age'] * 365
df.tail()
```

```python
df_above_30 = df[df['age']> 30]
df_above_30.head()
```

```python
df_18 = df[ (df["age"] == 18)  &  (df["Sex"]=="Female") ]
df_18.head()
```

```python
sorted_df = df.sort_values(by='age',  ascending = True)
sorted_df.head()
```

```python
sorted_df = df.sort_values(by="age", ascending = False)
sorted_df.head(5)
```

```python
df.groupby("Sex").size()
```

```python
df.drop_duplicates(inplace=True)
```

```python
gender_counts = df["Sex"].value_counts()
plt.figure(figsize= (8, 6))
plt.pie(gender_counts, labels = gender_counts.index, autopct ='%1.1f%%' , startangle = 140)
plt.title("Distribution by Gender")
plt.axis("equal")
plt.show()
```

```python
education_counts = df["education"].value_counts()
plt.figure(figsize= (8, 6))
plt.pie(gender_counts, labels = gender_counts.index, autopct ='%1.1f%%' , startangle = 140)
plt.title("Distribution by Education")
plt.axis("equal")
plt.show()
```

```python
education_counts = df["education"].value_counts()
plt.figure(figsize= (10, 6))
plt.bar(education_counts.index, education_counts.values, color="skyblue")
plt.xlabel("Education")
plt.ylabel("Count")
plt.title("Distribution by Education")
plt.xticks(rotation=45)
plt.show()
```
