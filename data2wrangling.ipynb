```python
# Import pandas package
import pandas as pd
# Assign data
data = {'Name': ['Jai', 'Princi', 'Gaurav','Anuj','Ravi','Natasha', 'Riya'],'Age':[17, 17, 18, 17, 18, 17, 17],'Gender': ['M', 'F', 'M', 'M', 'M', 'F', 'F'],'Blood_Glucose':[120, 119, 'NaN', 118, 122, 'NaN', 118]}
# Convert into DataFrame
df = pd.DataFrame(data)
count = df['Blood_Glucose'].count()
# Display data
print(count)
df

# Compute average
c = avg = 0
for ele in df['Blood_Glucose']:
    if str(ele).isnumeric():
        c += 1
        avg += ele
avg /= c

# Replace missing values and convert to numeric
df['Blood_Glucose'] = df['Blood_Glucose'].replace(to_replace="NaN", value=avg).astype(float) # Convert the column to numeric type after replacing NaN

# Display data
df

# Categorize gender
df['Gender'] = df['Gender'].map({'M': 0,
                                 'F': 1, }).astype(float)

# Display data
df

# Filter top Blood_Glucose
df = df[df['Blood_Glucose'] >= 120].copy()

# Remove age column from filtered DataFrame
df.drop('Age', axis=1, inplace=True)

# Display data
df
```
