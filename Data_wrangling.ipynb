```python
# import pandas package
import pandas as pd
```

```python
# assign data
data = {'Name' :['Jai', 'Princi', 'Gaurav', 'Anuj', 'Ravi', 'Natasha', 'Riya'],
        'Age': [17, 17, 18, 17, 18, 17, 17],
        'Gender': ['M', 'F', 'M', 'M', 'M', 'F', 'F'],
        'Blood_Glucose': [120, 119, 'NaN', 118, 122, 'NaN', 118]
        }
```

```python
# convert into data frame
df = pd.DataFrame(data)
```

```python
# display data
df
```

```python
df.head()
```

```python
# compute average
# Compute average
c = avg = 0
for ele in df['Blood_Glucose']:
    if str(ele).isnumeric():
        c += 1
        avg += ele
avg /= c
```

```python
# replace missing values
df = df.replace(to_replace="NaN", value=avg)
df
```

```python
# filter top blood_glucose
df = df[df['Blood_Glucose'] >= 120].copy()
df
```
