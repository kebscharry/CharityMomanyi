## **Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Set style for seaborn
sns.set(style="whitegrid")
```

## **Load and Explore Data**

```python
# Load the dataset
filepath = '/content/Yemen Cholera Outbreak Epidemiology Data - Data_Country_Level.csv'
df = pd.read_csv(filepath)

# Display the first few rows
df.head()

# Display summary statistics
df.describe()

# Check for missing values
df.isnull().sum()
```

## **Visualizations**

```python
# Distribution of cases
plt.figure(figsize=(10, 6))
sns.histplot(df['Cases'], kde=True)
plt.title('Distribution of Cholera Cases')
plt.xlabel('Number of Cases')
plt.ylabel('Frequency')
plt.show()

# Distribution of deaths
plt.figure(figsize=(10, 6))
sns.histplot(df['Deaths'], kde=True)
plt.title('Distribution of Cholera Deaths')
plt.xlabel('Number of Deaths')
plt.ylabel('Frequency')
plt.show()

# CFR and Attack Rate over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Date', y='CFR (%)', label='Case Fatality Rate (%)')
sns.lineplot(data=df, x='Date', y='Attack Rate (per 1000)', label='Attack Rate (per 1000)')
plt.title('CFR and Attack Rate Over Time')
plt.xticks(rotation=45)
plt.show()
```
